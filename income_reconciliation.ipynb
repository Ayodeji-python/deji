```python
#trying to append appropriate dates to the details of each transaction
# Define the search strings
search_strings_2022 = ['Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
search_strings_2023 = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']

# Create a regular expression pattern
pattern_2022 = '|'.join(search_strings_2022)
pattern_2023 = '|'.join(search_strings_2023)

# Specify the year to concatenate
year_to_concat_2022 = '2022'
year_to_concat_2023 = '2023'

# Filter 'Date' column for values containing any of the search strings for 2022
filtered_New_Df_2022 = new_df[new_df['New_Date'].str.contains(pattern_2022)]

# Concatenate the year to the existing date
filtered_New_Df_2022['New_Date'] = filtered_New_Df_2022['New_Date'] + ' ' + year_to_concat_2022

# Filter 'Date' column for values containing any of the search strings for 2023
filtered_New_Df_2023 = new_df[new_df['New_Date'].str.contains(pattern_2023)]

# Concatenate the year to the existing date
filtered_New_Df_2023['New_Date'] = filtered_New_Df_2023['New_Date'] + ' ' + year_to_concat_2023


# Concatenate the modified DataFrames
concatenated_df = pd.concat([filtered_New_Df_2022, filtered_New_Df_2023])

concatenated_df
```
